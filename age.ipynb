```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv','rt',encoding='utf-8-sig')
data = csv.reader(f)
next(data)
result = []
city = str(input("도시를 입력하세요: "))
for row in data:
    if city in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(",","")))

print(result)
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```
